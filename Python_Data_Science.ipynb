Over-riding __init__ Method

```python
class Vehicle:
    def __init__(self,milage,cost):
        self.milage = milage
        self.cost = cost
        
    def vehicle_details(self):
        print("I am a vehicle")
        print("Milage of vehicle is : ",self.milage)
        print("Cost of vehicle is : ",self.cost)
```

```python
class Car(Vehicle):
    def __init__(self,milage,cost,types,hp):
        super().__init__(milage,cost)
        self.types = types
        self.hp = hp
        
    def show_car_details(self):
        print("I am a car")
        print("Number of types are : ",self.types)
        print("Value of horse power is : ",self.hp)
        
    def show_all_details(self):
        print("Milage of vehicle is : ",self.milage)
        print("Cost of vehicle is : ",self.cost)
        print("Number of types are : ",self.types)
        print("Value of horse power is : ",self.hp)
```

```python
c1 = Car(1200,1000,5,350)
```

```python
c1.vehicle_details()
```

```python
c1.show_car_details()
```

```python
c1.show_all_details()
```

Multiple Inheritance in Python

```python
class Parent1:
    def assign_string_one(self,str1):
        self.str1 = str1
    def show_string_one(self):
        return self.str1
```

```python
class Parent2:
    def assign_string_two(self,str2):
        self.str2 = str2
    def show_string_two(self):
        return self.str2
```

```python
class Derived(Parent1,Parent2):
    def assign_string_three(self,str3):
        self.str3 = str3
    def show_string_three(self):
        return self.str3
```

```python
d1 = Derived()
```

```python
d1.assign_string_one("I am string of parent 1")
```

```python
d1.assign_string_two("I am string of parent 2")
```

```python
d1.assign_string_three("I am string of parent 3")
```

```python
d1.show_string_one()
```

```python
d1.show_string_two()
```

```python
d1.show_string_three()
```

Multi-Lavel Inheritance in Python

```python
class Parent:
    def set_name(self,name):
        self.name = name
    def get_name(self):
        return self.name
```

```python
class Child(Parent):
    def set_age(self,age):
        self.age = age
    def get_age(self):
        return self.age
```

```python
class Grandchild(Child):
    def set_gender(self,gender):
        self.gender = gender
    def get_gender(self):
        return self.gender
```

```python
gc = Grandchild()
```

```python
gc.set_name("Shamsur Rahman")
```

```python
gc.set_gender("Male")
```
